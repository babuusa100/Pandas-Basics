# Missing value imputation:


```python
admit = pd.read_csv('C:\\Users\\BABU\\Downloads\\admit_missing.csv')
#to know the number of missing values in each variable
admit.isna().sum()
```

# Types of mannualimputations
a)ForwardFill

b)BackwardFill

c)Mean-Imputation

d)Median-Imputation

e)Mode-Imputation

```python
# now imputing missing values in gre variable
GRE = admit["GRE"]
GRE_ffill = GRE.fillna(GRE.ffill())# using the forward filling for imputation

admit["GRE"]= GRE_ffill
admit.isna().sum()
```

```python
# imputing SOP using the backwardfill
sop = admit["SOP"]
sop_fill = sop.fillna(sop.bfill())
admit["SOP"]=sop_fill
```

```python
#Mean imputation:
cgpa = admit["CGPA"]
cgpaf= cgpa.fillna(cgpa.mean())#for median imputation use .median instead
admit["CGPA"]=cgpaf
admit.isna().sum()
```

```python
#Mode-Imputation:
ur = admit["Univ_Rating"]
md = int(ur.mode())
urf = ur.fillna(md)
admit["Univ_Rating"] = urf
admit.isna().sum()
```
